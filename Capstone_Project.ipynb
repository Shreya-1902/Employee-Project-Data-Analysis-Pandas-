```python

# Task 1
import pandas as pd
import numpy as np

data = {}

project_df = pd.DataFrame(data)

project_df.to_csv("project.csv", index=False)
project_df.to_csv("employee.csv", index=False)
project_df.to_csv("seniority.csv", index=False)
project_df.to_csv("final.csv", index=False)


# Creating dictionary for Project DataFrame
project_data = {
    "ID": ["A001","A002","A003","A004","A005","A002","A005","A003","A001","A003","A001","A004","A004","A005"],
    "Project": ["Project 1","Project 2","Project 3","Project 4","Project 5",
                "Project 6","Project 7","Project 8","Project 9","Project 10",
                "Project 11","Project 12","Project 13","Project 14"],
    "Cost": [1002000,2000000,4500000,5500000,np.nan,
             680000,400000,350000,np.nan,300000,
             2000000,1000000,3000000,200000],
    "Status": ["Finished","Ongoing","Finished","Ongoing","Finished",
               "Failed","Finished","Failed","Ongoing","Finished",
               "Failed","Ongoing","Finished","Finished"]
}

# Convert dictionary to DataFrame
project_df = pd.DataFrame(project_data)

# Save to CSV file
project_df.to_csv("project.csv", index=False)

# Display DataFrame
project_df

# Creating dictionary for Employee DataFrame
employee_data = {
    "ID": ["A001","A002","A003","A004","A005"],
    "Name": ["John Alter","Alice Luxumberg","Tom Sabestine","Nina Adgra","Amy Johny"],
    "Gender": ["M","F","M","F","F"],
    "City": ["Paris","London","Berlin","Newyork","Madrid"],
    "Age": [25,27,29,31,30]
}

# Convert dictionary to DataFrame
employee_df = pd.DataFrame(employee_data)

# Save to CSV
employee_df.to_csv("employee.csv", index=False)

# Display DataFrame
employee_df

# Creating dictionary for Seniority DataFrame
seniority_data = {
    "ID": ["A001","A002","A003","A004","A005"],
    "Designation Level": [2,2,3,2,3]
}

# Convert dictionary to DataFrame
seniority_df = pd.DataFrame(seniority_data)

# Save to CSV
seniority_df.to_csv("seniority.csv", index=False)

# Display DataFrame
seniority_df


# Task 2
# Load project CSV file
project_df = pd.read_csv("project.csv")

# Initialize running total and count
total = 0
count = 0

# Loop through rows
for i in range(len(project_df)):
    
    # If cost is not missing
    if not pd.isna(project_df.loc[i, "Cost"]):
        total += project_df.loc[i, "Cost"]
        count += 1
    
    # If cost is missing, replace with running average
    else:
        project_df.loc[i, "Cost"] = total / count

# Save updated file
project_df.to_csv("project.csv", index=False)

# Display updated DataFrame
project_df


#Task 3
# Load employee CSV
employee_df = pd.read_csv("employee.csv")

# Split Name into First Name and Last Name
employee_df[["First Name", "Last Name"]] = employee_df["Name"].str.split(" ", expand=True)

# Drop original Name column
employee_df.drop("Name", axis=1, inplace=True)

# Save updated file
employee_df.to_csv("employee.csv", index=False)

# Display updated DataFrame
employee_df


# Task 4
# Reload all CSV files
project_df = pd.read_csv("project.csv")
employee_df = pd.read_csv("employee.csv")
seniority_df = pd.read_csv("seniority.csv")

# Merge project and employee
final = pd.merge(project_df, employee_df, on="ID")

# Merge with seniority
final = pd.merge(final, seniority_df, on="ID")

# Save final DataFrame
final.to_csv("final.csv", index=False)

# Display final DataFrame
final


# Task 5
# Add Bonus column (5% for Finished projects)
final["Bonus"] = np.where(final["Status"] == "Finished",
                          final["Cost"] * 0.05,
                          0)

final


# Task 6
# Increase designation level by 1 for Failed projects
final.loc[final["Status"] == "Failed", "Designation Level"] += 1

# Remove records where designation level > 4
final = final[final["Designation Level"] <= 4]

final


# Task 7
# Add title based on Gender
final["First Name"] = np.where(final["Gender"] == "M",
                               "Mr. " + final["First Name"],
                               "Mrs. " + final["First Name"])

# Drop Gender column
final.drop("Gender", axis=1, inplace=True)

final

# Task 8
# Increase designation level by 1 where Age > 29
final.loc[final["Age"] > 29, "Designation Level"] += 1

final


# Task 9
# Group by ID and First Name to calculate total cost
TotalProjCost = final.groupby(["ID", "First Name"])["Cost"].sum().reset_index()

# Rename column
TotalProjCost.rename(columns={"Cost": "Total cost"}, inplace=True)

#Display the new dataframe
print(TotalProjCost)


# Task 10
# Filter employees whose city contains letter "o"
filtered_employees = final[final["City"].str.contains("o", case=False)]

#Display the result
print(filtered_employees)
```
